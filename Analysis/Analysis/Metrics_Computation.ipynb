## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jiwer import wer, cer, compute_measures
import os
```

## WER and CER Functions

```python
# Function to add WER and CER columns
def add_wer_cer_columns(df):
    df['WER'] = df.apply(lambda row: wer(row['Gold Label'], row['Response']), axis=1)
    df['CER'] = df.apply(lambda row: cer(row['Gold Label'], row['Response']), axis=1)
    df['word_count'] = df['Gold Label'].apply(lambda x: len(str(x).split()))  # Add word count column
    return df

# Function to visualize WER and CER distributions
def plot_distribution(df, column, title, xlabel):
    plt.figure(figsize=(10, 6))
    plt.hist(df[column], bins=30, alpha=0.7, edgecolor='black')
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Function to plot WER and CER against Word Count
def plot_scatter(df, y_column, title, ylabel, color='blue'):
    plt.figure(figsize=(12, 6))
    plt.scatter(df['word_count'], df[y_column], alpha=0.6, color=color, label=y_column)
    plt.title(title, fontsize=16)
    plt.xlabel('Word Count', fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.grid(alpha=0.5)
    plt.legend()
    plt.show()

# Function for Box Plot
def plot_box(df, column, title, ylabel):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=column, data=df, palette='pastel')
    plt.title(title, fontsize=16)
    plt.ylabel(ylabel, fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Function for Violin Plot
def plot_violin(df, column, title, ylabel):
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=column, data=df, palette='muted')
    plt.title(title, fontsize=16)
    plt.ylabel(ylabel, fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Function for CDF Plot
def plot_cdf(df, column, title, xlabel):
    plt.figure(figsize=(10, 6))
    sns.ecdfplot(data=df, x=column, color='blue')
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel('Cumulative Probability', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

## URDU Evaluation

```python
# Process Urdu Dataset
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

urdu_file_path = 'urdu_inferenced_dataset.csv'
urdu_df = pd.read_csv(urdu_file_path)
urdu_df = add_wer_cer_columns(urdu_df)

# Save updated file
urdu_output_file = 'urdu_inferenced_dataset_with_metrics.csv'
urdu_df.to_csv(urdu_output_file, index=False)
print("Urdu dataset updated and saved.")

# Visualize WER and CER distributions for Urdu
plot_distribution(urdu_df, 'WER', 'Distribution of WER (Urdu)', 'WER')
plot_distribution(urdu_df, 'CER', 'Distribution of CER (Urdu)', 'CER')

# Scatter plots of WER and CER against Word Count for Urdu
plot_scatter(urdu_df, 'WER', 'WER vs Word Count (Urdu)', 'WER')
plot_scatter(urdu_df, 'CER', 'CER vs Word Count (Urdu)', 'CER', color='orange')

# Box Plot of WER and CER
plot_box(urdu_df, 'WER', 'Box Plot of WER (Urdu)', 'WER')
plot_box(urdu_df, 'CER', 'Box Plot of CER (Urdu)', 'CER')

# Violin Plot of WER and CER
plot_violin(urdu_df, 'WER', 'Violin Plot of WER (Urdu)', 'WER')
plot_violin(urdu_df, 'CER', 'Violin Plot of CER (Urdu)', 'CER')

# CDF Plot of WER and CER
plot_cdf(urdu_df, 'WER', 'CDF of WER (Urdu)', 'WER')
plot_cdf(urdu_df, 'CER', 'CDF of CER (Urdu)', 'CER')
```
